# EDA

```python
#Checking distribution of cgpa & iq

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].hist(df['iq'], label='IQ')
ax[0].set_xlabel('IQ')
ax[0].set_ylabel('Density')

ax[1].hist(df['cgpa'], label='CGPA')
ax[1].set_xlabel('CGPA')
ax[1].set_ylabel('Density')

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(df, x='cgpa', hue='placement', bins=30, edgecolor='black', kde=False, ax=ax[0])
ax[0].set_xlabel('CGPA')
ax[0].set_ylabel('Density')

sns.histplot(df, x='iq', hue='placement', bins=30, edgecolor='black', kde=False, ax=ax[1])
ax[1].set_xlabel('IQ')
ax[1].set_ylabel('Density')

plt.tight_layout()

plt.show()
```

Looking at the above and below graph, there is a clear pattern - high cgpa > placement is 1 (with some mixture in the middle)

```python
sns.scatterplot(data=df, x='cgpa', y='iq', hue='placement')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('CGPA vs IQ')
plt.show()
```

```python
sns.boxplot(data=df, x='placement', y='cgpa')
plt.xlabel('Placement')
plt.ylabel('CGPA')
plt.title('CGPA by Placement')
plt.show()

sns.boxplot(data=df, x='placement', y='iq')
plt.xlabel('Placement')
plt.ylabel('IQ')
plt.title('IQ by Placement')
plt.show()
```

```python
sns.pairplot(df, hue='placement')
plt.show()
```

```python
sns.countplot(data=df, x='placement')
plt.xlabel('Placement')
plt.ylabel('Count')
plt.show()
```

There is no multicollinearity between iq and placement

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='BuGn', fmt=".2f")
plt.show()
```
